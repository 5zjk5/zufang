# 行政区划分

```python
def region(s):
    if '宝安' in s: 
        return '宝安'
    if '南山' in s: 
        return '南山'
    if '龙华' in s: 
        return '龙华'
    if '福田' in s: 
        return '福田'
    if '龙岗' in s: 
        return '龙岗'
    if '罗湖' in s: 
        return '罗湖'
    if '布吉' in s: 
        return '布吉'
    if '光明' in s: 
        return '光明'
    if '坪山' in s: 
        return '坪山'
    if '盐田' in s: 
        return '盐田'
    if '大鹏新区' in s: 
        return '大鹏新区'

df['行政区'] = df['位置'].map(region)
df.head()
```

# 每个行政区价格分布

```python
# 把【价格】元/月 去掉，并转换为整型
df['价格'] = df['价格'].str.replace('元/月','')
df['价格'] = df['价格'].astype(int)

# 区间价格划分
df['price'] = pd.cut(df['价格'],bins=[0,500,1000,1500,2000,2500])
df['price'] = df['price'].astype(str)
```

# 根据设施筛选

```python
# 删除设施为空的数据
df = df.dropna()

# 选出含有【热水器】【洗衣机】【空调】【宽带】【卫生间】【床】的房源
def tj1(s):
    if ('热水器' in s) and (
            '洗衣机' in s) and (
            '空调' in s) and (
            '宽带' in s) and (
            '卫生间' in s) and (
            '床' in s):
         return True
    else:
        return False
    
df['tj1'] = df['设施'].map(tj1)
df = df[df['tj1'] == True]
df.shape[0]
```

# 筛选是否含有地铁

```python
# [概况] 是否含有地铁关键字
def train(s):
    if '地铁' in s:
        return True
    else:
        False
    
df['概况'] = df['概况'].map(train)
df = df[df['概况'] == True]
df.shape[0]
```

```python
# 标签中是否含有【5号线】
def train(s):
    if '号线' in s:
        return True
    else:
        False
    
df['标签'] = df['标签'].map(train)
df = df[df['标签'] == True]
df.shape[0]
```

# 筛选出1500以下的

```python
df = df[(df['price'] == '(0.0, 500.0]') | 
      (df['price'] == '(500.0, 1000.0]') |
      (df['price'] == '(1000.0, 1500.0]')]
df.shape[0]
```

```python
df['行政区'].value_counts()
```

# 根据实际筛选出距离合适的区

```python
df = df[(df['行政区'] == '宝安') | (df['行政区'] == '南山')]
df.shape[0]
```
